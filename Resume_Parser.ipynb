```python
from pyresparser import ResumeParser
from docx import Document

data = ResumeParser('/content/Ishank_CV.pdf').get_extracted_data()

doc = Document()

doc.add_heading('Resume Information', level=1)

def add_info_with_subheading(subheading, info):
    doc.add_heading(subheading, level=2)
    if info:
        for item in info:
            doc.add_paragraph(f"- {item}")

add_info_with_subheading('Name', [data['name']])
add_info_with_subheading('Email', [data['email']])
add_info_with_subheading('Skills', data['skills'])
add_info_with_subheading('Mobile Number', [data['mobile_number']])
add_info_with_subheading('College', [data['college_name']])
add_info_with_subheading('Degree', [data['degree']])
add_info_with_subheading('Designation', [data['designation']])

add_info_with_subheading('Work Experience', data['experience'])
add_info_with_subheading('Company Names', data['company_names'])

doc.add_heading('Additional Information', level=2)
doc.add_paragraph(f"Number of Pages: {data['no_of_pages']}")
doc.add_paragraph(f"Total Experience: {data['total_experience']}")

doc.save('resume_info.docx')
```
